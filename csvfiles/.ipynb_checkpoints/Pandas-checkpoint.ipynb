```python
#joins 
dataframe1 = pd.DataFrame({
    'employee': ['abc', 'xyz', 'mno'],
    'age': ['7653', '3567', '345678']
})
dataframe2 = pd.DataFrame({
    'employee': ['abc', 'xyz', 'pqr'],
    'salary': ['234', '3456', '775453']
})
dataframe3 = pd.merge(dataframe1, dataframe2, on='employee')
dataframe3
```

```python
dataframe3 = pd.merge(dataframe1, dataframe2, on='employee', how='outer')
dataframe3
```

```python
#left
dataframe3 = pd.merge(dataframe1, dataframe2, on='employee', how='left')
dataframe3
```

```python
#right
dataframe3 = pd.merge(dataframe1, dataframe2, on='employee', how='right')
dataframe3
```

```python
#shifting and logging
hdq = pd.read_csv('HistoricalQuotes.csv', index_col='date')
hdq.shift(1).head()
hdq.shift(-1).head()
```

```python
#calculate the stock price delta
hdq['volume'] = hdq['low'].shift(1)
hdq.head()
```

```python
hdq['delta'] = hdq['open'] - hdq['close']
hdq.head()
```

```python
hdq['Week Returns'] = ((hdq['close'] - hdq['close'].shift(7))/hdq['close'])*100
hdq
```

```python
#write data to csv
hdq.to_csv('hdq.csv')
hdq.to_csv('hdq_specs.csv', columns = ['delta', 'Week Returns'])
```
